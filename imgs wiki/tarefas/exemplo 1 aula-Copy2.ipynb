```python
import scipy as sp
import matplotlib.pyplot as plt
from ipywidgets import interact
%matplotlib inline
```

```python
wn = 2*2*sp.pi
wf = sp.linspace(0,2*wn,1000)
f0 = 1
```

```python
def f(ksi):
    x  = f0/((((wn**2)-(wf**2))**2+((2*ksi*wn*wf)**2))**0.5)
    plt.plot(wf/wn,x)
```

```python
interact(f, ksi = 0.1)
```

```python
def v(ksi):
    o  = sp.arctan(2*ksi*wn*wf/((wn**2)-(wf**2)))
    plt.plot(wf/wn,o)
```

```python
interact(v, ksi = 0.1)
```
