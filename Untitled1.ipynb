```python

import scipy as sp
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
%matplotlib inline 
```

```python
m = 2
k = 10e5
ksi = 0.5
c = 2*ksi*(m*k)**0.5
n = 2*ksi
omega = sp.linspace(0, 2*(k/m)**0.5, 1000)
```

```python
vw = []
vtc = []
vt = []
vwc = []
for O in omega:
    x = 1/(((-O**2)*m)+1.j*O*c+k)
    vw.append(sp.absolute(x))
for O in omega:
    y = 1/(((-O**2)*m)+(1.j*n+1)*k)
    vtc.append(sp.absolute(y))
ksi = 0.05
c = 2*ksi*(m*k)**0.5
n = 2*ksi
for O in omega:
    z = 1/(((-O**2)*m)+1.j*O*c+k)
    vt.append(sp.absolute(z))
for O in omega:
    w = 1/(((-O**2)*m)+(1.j*n+1)*k)
    vwc.append(sp.absolute(w))
```

```python
plt.plot(omega,vwc,omega,vt)
```
